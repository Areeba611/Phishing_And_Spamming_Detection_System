```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer=RegexpTokenizer(r'[A-Za-z]+')
```

```python
df.URL[0]
```

```python
tokenizer.tokenize(df.URL[0])
```

```python
df['text_tokenized'] =df.URL.map(lambda t: tokenizer.tokenize(t))
```

```python
df.head()
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
```

```python
df['text_stemmed']= df['text_tokenized'].map(lambda l: [stemmer.stem(word) for word in l])
```

```python
df.head()
```

```python
df['text'] = df['text_stemmed'].map(lambda l:' '.join(l))
```

```python
df.head()
```

```python
good_sites = df[df.Label == 'good']
bad_sites = df[df.Label == 'bad' ]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
features = cv.fit_transform(df.text)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, df.Label , test_size=0.3)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
l_model= LogisticRegression()
```

```python
l_model.fit(x_train, y_train)
```

```python
l_model.score(x_test,y_test)
```

```python
l_model.score(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
print('\n Classification Report\n')
print(classification_report(l_model.predict(x_test), y_test, target_names = ['Bad' , 'Good']))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
con_mat= pd.DataFrame(confusion_matrix(l_model.predict(x_test), y_test),
            columns=['Predicted:Bad','Predicted:Good'],index = ['Actual:Bad', 'Actual:Good'])
```

```python
import seaborn as sns
```

```python
print('\nConfusion Matrix')
plt.figure(figsize=(6,4))
sns.heatmap(con_mat,annot=True,fmt='d',cmap="YlGnBu")
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)
```
